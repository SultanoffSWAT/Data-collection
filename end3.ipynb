```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
df = pd.read_csv('/kaggle/input/amazon-sales-dataset/amazon.csv')
df.columns
```

```python
df.head(5)
```

```python
df['discounted_price'] = df['discounted_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
df['actual_price'] = df['actual_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
df[['discounted_price', 'actual_price']].head()
```

```python
df['discount_percentage'] = df['discount_percentage'].replace({'%': ''}, regex=True).astype(int)
df['discount_percentage'].head()
```

```python
df.info()
```

```python
# df.describe(include='all')
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

columns_to_plot = ['discounted_price', 'actual_price', 'discount_percentage']
fig, axes = plt.subplots(ncols=3, figsize=(18, 12))
axes = axes.flatten()

for i, column in enumerate(columns_to_plot):
    sns.boxplot(x=df[column], ax=axes[i])
    axes[i].set_title(f'Boxplot of {column} with Outliers')
    axes[i].set_xlabel(column)
if 5 < len(axes):
    fig.delaxes(axes[-1])
plt.tight_layout()
plt.show()
```
